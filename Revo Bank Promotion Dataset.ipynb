# Add Promo Sensitive Customers Identifier Column

Now we create a new column to identify promo-sensitive customers (more than 50% of their total sales in the revoshop are attributed to the promo)

So the formula of total sales is AVG_TXN_AMT_LTM * TXN_CNT_LTM
and the formula of total sales attributed to the promo is  AVG_PROMO_TXN_AMT_LTM * PROMO_TXN_CNT_LTM

Then if total sales attributed to promo/total sales is more than 50% (0.5) then the customers is promo sensitive

First we make a new copy from the cleaned dataset

```python
df_dce = df_dc.copy()
```

Then we create 3 new columns (TOTAL_SALES, TOTAL_SALES_PROMO, and PROMO_ATTRIBUTED_SALES)

```python
df_dce['TOTAL_SALES'] = df_dce['AVG_TXN_AMT_LTM'] * df_dce['TXN_CNT_LTM']
```

```python
df_dce['TOTAL_SALES_PROMO'] = df_dce['AVG_PROMO_TXN_AMT_LTM'] * df_dce['PROMO_TXN_CNT_LTM']
```

```python
df_dce['PROMO_ATTRIBUTED_SALES'] = df_dce['TOTAL_SALES_PROMO']/df_dce['TOTAL_SALES']
```

```python
df_dce
```

Then we create PROMO_SENSITIVE columns with conditions

```python
conditions = [
    (df_dce['PROMO_ATTRIBUTED_SALES'] >= 0.5),
    (df_dce['PROMO_ATTRIBUTED_SALES'] < 0.5)
    ]
values = ['YES', 'NO']
df_dce['PROMO_SENSITIVE'] = np.select(conditions, values)
```

```python
df_dce
```

```python
for col in df_dce.columns:
  print (col)
  print(df_dce[col].value_counts())
  print('\n')
```

# Elbow Method

```python
distortions = []
K = range(1,11)
for k in K:
    kmeanModel = KMeans(k, random_state = 1000, n_init='auto')
    kmeanModel.fit(df_clustering3)
    distortions.append(kmeanModel.inertia_)
print(distortions)

plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'b*-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

There are several potential elbows here, such as k=3 and k=4

# Silhouette Method

Untuk cluster k =3

```python
cluster_model = KMeans(n_clusters = 3, random_state = 1000, n_init='auto')
cluster_model.fit(df_clustering3)

sil = SilhouetteVisualizer(cluster_model, is_fitted = True)
sil.fit(df_clustering3)
sil.finalize()
sil.silhouette_score_
```

```python
K = range(2,11)
max_K = max(K)
fig, ax = plt.subplots(int(np.ceil(max_K/2)), 2, figsize = (15,30))

for n_clusters in K:
  kmeanModel = KMeans(n_clusters, random_state = 1000, n_init='auto')

  q, mod = divmod(n_clusters,2)
  sil = SilhouetteVisualizer(kmeanModel, is_fitted = False, ax = ax[q-1][mod])
  sil.fit(df_clustering3)
  sil.finalize()
  print(f"For k={n_clusters}, the average silhouette score is {sil.silhouette_score_}")
```

1. By elbow method, k=2 or k=3
2. By highest silhouette score, k=2
3. k=3 looks okay with minimal negative coef values
4. We're gonna continue with 3 cluster
